```python
# CELLULE 5 - Envoyer automatiquement vers le canal bangbros
import os, requests

def send_to_channel(folder, bot_token, channel):
    if not channel:
        print('CHANNEL non defini - lance la cellule 3b')
        return
    exts = ('.mp4', '.mkv', '.avi', '.mov', '.webm')
    videos = sorted([f for f in os.listdir(folder) if f.lower().endswith(exts)])
    if not videos:
        print('Aucune video dans', folder)
        return
    print(f'{len(videos)} video(s) -> canal {channel}')
    ok = 0
    for fname in videos:
        path = os.path.join(folder, fname)
        size_mb = os.path.getsize(path) / 1024 / 1024
        if size_mb > 2000:
            print(f'[IGNORE] {fname} ({size_mb:.0f}MB > 2GB)')
            continue
        print(f'Envoi {fname} ({size_mb:.1f}MB)...')
        with open(path, 'rb') as f:
            r = requests.post(
                f'https://api.telegram.org/bot{bot_token}/sendVideo',
                data={'chat_id': channel, 'caption': fname[:1000], 'supports_streaming': 'true'},
                files={'video': (fname, f, 'video/mp4')},
                timeout=600
            )
        if r.status_code == 200:
            print('  OK')
            ok += 1
        else:
            err = r.json()
            print(f'  Erreur {r.status_code}: {err.get("description",r.text[:100])}')
    print(f'Termine: {ok}/{len(videos)} envoyes')

send_to_channel(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
```

```python
# CELLULE 6 - Agent Hermes (optionnel)
# Laisse tourner pour recevoir des ordres depuis @TGBOX_HQbot
import time, threading, requests as req

BASE = MASTER_URL.rstrip('/')

def register():
    try:
        r = req.post(f'{BASE}/report/{AGENT_ID}',
                     json={'status': 'ready', 'agent_id': AGENT_ID}, timeout=8)
        return r.status_code == 200
    except Exception as e:
        print('Mastermind inaccessible:', e)
        return False

def poll_order():
    try:
        r = req.get(f'{BASE}/poll/{AGENT_ID}', timeout=8)
        if r.status_code == 200:
            return r.json().get('order')
    except Exception:
        pass
    return None

def report(status):
    try:
        req.post(f'{BASE}/report/{AGENT_ID}',
                 json={'status': status, 'agent_id': AGENT_ID}, timeout=5)
    except Exception:
        pass

def execute(order):
    parts = order.strip().split(None, 1)
    cmd_name = parts[0].lower()
    args = parts[1] if len(parts) > 1 else ''
    print(f'[Agent] {order}')
    report(f'running: {order[:50]}')
    if cmd_name == 'download' and args:
        r = subprocess.run([
            sys.executable, '-m', 'zeusdl',
            '--cookies', COOKIES_PATH,
            '-f', f'bestvideo[height<={QUALITY}]+bestaudio/best[height<={QUALITY}]/best',
            '--merge-output-format', 'mp4',
            '-o', OUTPUT_DIR + '/%(title)s.%(ext)s',
            '--concurrent-fragments', str(WORKERS),
            args
        ])
        report('done' if r.returncode == 0 else 'error')
    elif cmd_name == 'push' and 'telegram' in args:
        send_to_channel(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
        report('done')
    else:
        print(f'[Agent] Inconnu: {cmd_name}')
        report('ready')

if register():
    print(f'Agent {AGENT_ID} connecte')
    print(f'/send {AGENT_ID} download https://site-ma.bangbros.com/scenes?addon=5971')
    print(f'/send {AGENT_ID} push telegram')
    while True:
        order = poll_order()
        if order:
            threading.Thread(target=execute, args=(order,), daemon=True).start()
        time.sleep(3)
else:
    print('Lance cellules 4 et 5 manuellement.')
```
