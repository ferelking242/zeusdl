```python
# ================================================================
# CELLULE 3 - Connexion BangBros (email + mot de passe)
# ================================================================
import json
import time
import requests

def bangbros_login(email, password, cookies_path):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/124.0 Safari/537.36',
        'Accept': 'application/json',
        'Content-Type': 'application/json',
        'Origin': 'https://www.bangbros.com',
        'Referer': 'https://www.bangbros.com/',
        'X-Site-Slug': 'bangbros',
    }
    payload = {'username': email, 'password': password}
    resp = requests.post(
        'https://site-api.project1service.com/api/account/login',
        json=payload, headers=headers, timeout=20
    )
    data = resp.json()
    if resp.status_code != 200 or not data.get('access_token') and not data.get('data', {}).get('access_token'):
        print('Reponse brute:', json.dumps(data, indent=2)[:500])
        raise Exception('Login echoue : ' + str(data.get('message') or data.get('error') or resp.status_code))

    token = data.get('access_token') or data.get('data', {}).get('access_token') or data.get('token')
    instance = data.get('instance_token') or data.get('data', {}).get('instance_token') or ''
    exp = int(time.time()) + 30 * 24 * 3600

    lines = ['# Netscape HTTP Cookie File']
    for domain in ['.bangbros.com', '.site-ma.bangbros.com', '.project1service.com', '.site-api.project1service.com']:
        lines.append(f'{domain}\tTRUE\t/\tFALSE\t{exp}\taccess_token_ma\t{token}')
        if instance:
            lines.append(f'{domain}\tTRUE\t/\tFALSE\t{exp}\tinstance_token\t{instance}')

    os.makedirs(os.path.dirname(os.path.abspath(cookies_path)), exist_ok=True)
    with open(cookies_path, 'w') as f:
        f.write('\n'.join(lines) + '\n')
    os.chmod(cookies_path, 0o600)
    return cookies_path

cookies_file = bangbros_login(BB_EMAIL, BB_PASSWORD, COOKIES_PATH)
print('Connexion OK - cookies sauvegardes dans', cookies_file)
```

```python
# ================================================================
# CELLULE 5 - Envoyer les videos vers Telegram
# ================================================================
import os
import requests

def send_to_telegram(folder, bot_token, channel):
    if not channel:
        print('CHANNEL vide - mets ton channel ID dans la cellule 2')
        return
    extensions = ('.mp4', '.mkv', '.avi', '.mov', '.webm')
    files = [f for f in os.listdir(folder) if f.lower().endswith(extensions)]
    if not files:
        print('Aucune video trouvee dans', folder)
        return
    print(f'{len(files)} video(s) a envoyer vers {channel}')
    for fname in files:
        path = os.path.join(folder, fname)
        size_mb = os.path.getsize(path) / 1024 / 1024
        if size_mb > 2000:
            print(f'Ignore (trop gros {size_mb:.0f}MB): {fname}')
            continue
        print(f'Envoi: {fname} ({size_mb:.1f} MB)...')
        with open(path, 'rb') as f:
            r = requests.post(
                f'https://api.telegram.org/bot{bot_token}/sendVideo',
                data={'chat_id': channel, 'caption': fname[:1000]},
                files={'video': (fname, f, 'video/mp4')},
                timeout=300
            )
        if r.status_code == 200:
            print(f'  OK')
        else:
            print(f'  Erreur: {r.text[:200]}')

send_to_telegram(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
```

```python
# ================================================================
# CELLULE 6 - Connecter cet agent au Mastermind Replit (optionnel)
# Lance cette cellule si tu veux controler Colab depuis @TGBOX_HQbot
# ================================================================
import time
import json
import threading
import requests

class SimpleAgent:
    def __init__(self, agent_id, master_url, cookies_path, output_dir, quality, workers, bot_token, channel):
        self.agent_id = agent_id
        self.master_url = master_url.rstrip('/')
        self.cookies_path = cookies_path
        self.output_dir = output_dir
        self.quality = quality
        self.workers = workers
        self.bot_token = bot_token
        self.channel = channel
        self.running = True

    def register(self):
        try:
            r = requests.post(f'{self.master_url}/register',
                json={'agent_id': self.agent_id, 'status': 'ready'}, timeout=5)
            return r.status_code == 200
        except Exception as e:
            print('Mastermind inaccessible:', e)
            return False

    def poll(self):
        try:
            r = requests.get(f'{self.master_url}/orders/{self.agent_id}', timeout=5)
            if r.status_code == 200:
                return r.json().get('orders', [])
        except Exception:
            pass
        return []

    def ack(self, order_id):
        try:
            requests.post(f'{self.master_url}/ack',
                json={'agent_id': self.agent_id, 'order_id': order_id}, timeout=5)
        except Exception:
            pass

    def execute(self, order):
        import subprocess, os
        print(f'[Agent] Ordre: {order}')
        parts = order.strip().split(None, 1)
        cmd = parts[0].lower()
        args = parts[1] if len(parts) > 1 else ''

        if cmd == 'download' and args:
            proc = subprocess.run([
                'yt-dlp', '--cookies', self.cookies_path,
                '-f', f'bestvideo[height<={self.quality}]+bestaudio/best[height<={self.quality}]',
                '--merge-output-format', 'mp4',
                '-o', os.path.join(self.output_dir, '%(title)s.%(ext)s'),
                '--concurrent-fragments', str(self.workers),
                args
            ])
            print('[Agent] Telechargement', 'OK' if proc.returncode == 0 else 'ERREUR')

        elif cmd == 'push' and 'telegram' in args:
            send_to_telegram(self.output_dir, self.bot_token, self.channel)

        elif cmd == 'login':
            bangbros_login(BB_EMAIL, BB_PASSWORD, self.cookies_path)
            print('[Agent] Re-connexion BangBros OK')

        else:
            print(f'[Agent] Commande inconnue: {cmd}')

    def start(self):
        ok = self.register()
        if not ok:
            print('Impossible de joindre le Mastermind:', self.master_url)
            print('Lance les cellules 4 et 5 directement a la place.')
            return
        print(f'Agent {self.agent_id} connecte au Mastermind')
        print('Depuis @TGBOX_HQbot tu peux envoyer :')
        print(f'  /agents')
        print(f'  /send {self.agent_id} download URL')
        print(f'  /send {self.agent_id} push telegram')
        print('En attente d\'ordres... (laisse cette cellule tourner)')
        while self.running:
            try:
                orders = self.poll()
                for o in orders:
                    oid = o.get('id')
                    text = o.get('command') or o.get('order') or ''
                    if text:
                        threading.Thread(target=self.execute, args=(text,), daemon=True).start()
                    if oid:
                        self.ack(oid)
            except Exception as e:
                print('[Agent] Erreur poll:', e)
            time.sleep(3)

agent = SimpleAgent(
    agent_id=AGENT_ID,
    master_url=MASTER_URL,
    cookies_path=COOKIES_PATH,
    output_dir=OUTPUT_DIR,
    quality=QUALITY,
    workers=WORKERS,
    bot_token=BOT_TOKEN,
    channel=CHANNEL
)
agent.start()
```
