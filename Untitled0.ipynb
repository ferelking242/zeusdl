```python
# CELLULE 5 - Envoyer les videos vers Telegram
import os, requests

def send_to_telegram(folder, bot_token, channel):
    if not channel:
        print('CHANNEL vide - mets ton channel ID dans la cellule 2')
        print('Ex: CHANNEL = "-1001234567890"')
        return
    exts = ('.mp4', '.mkv', '.avi', '.mov', '.webm')
    videos = [f for f in os.listdir(folder) if f.lower().endswith(exts)]
    if not videos:
        print('Aucune video dans', folder)
        return
    print(f'{len(videos)} video(s) a envoyer vers {channel}')
    for fname in videos:
        path = os.path.join(folder, fname)
        size_mb = os.path.getsize(path) / 1024 / 1024
        if size_mb > 2000:
            print(f'Ignore (trop gros {size_mb:.0f}MB): {fname}')
            continue
        print(f'Envoi {fname} ({size_mb:.1f}MB)...')
        with open(path, 'rb') as f:
            r = requests.post(
                f'https://api.telegram.org/bot{bot_token}/sendVideo',
                data={'chat_id': channel, 'caption': fname[:1000]},
                files={'video': (fname, f, 'video/mp4')},
                timeout=300
            )
        print('  OK' if r.status_code == 200 else f'  Erreur {r.status_code}: {r.text[:150]}')

send_to_telegram(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
```

```python
# CELLULE 6 - Agent Hermes (optionnel)
# Connecte ce Colab au Mastermind sur Replit
# Controle ensuite depuis @TGBOX_HQbot
import time, threading, subprocess, requests as req

# Endpoints corrects du Mastermind :
#   GET  /poll/{agent_id}        -> {"order": "..."}  (None si vide)
#   POST /report/{agent_id}      -> enregistrement + status

BASE = MASTER_URL.rstrip('/')

def register():
    try:
        r = req.post(f'{BASE}/report/{AGENT_ID}',
                     json={'status': 'ready', 'agent_id': AGENT_ID},
                     timeout=8)
        return r.status_code == 200
    except Exception as e:
        print('Mastermind inaccessible:', e)
        return False

def poll_order():
    try:
        r = req.get(f'{BASE}/poll/{AGENT_ID}', timeout=8)
        if r.status_code == 200:
            return r.json().get('order')
    except Exception:
        pass
    return None

def report_status(status):
    try:
        req.post(f'{BASE}/report/{AGENT_ID}',
                 json={'status': status, 'agent_id': AGENT_ID},
                 timeout=5)
    except Exception:
        pass

def execute(order):
    parts = order.strip().split(None, 1)
    cmd_name = parts[0].lower()
    args = parts[1] if len(parts) > 1 else ''
    print(f'[Agent] Execution: {order}')
    report_status(f'running: {order[:50]}')

    if cmd_name == 'download' and args:
        result = subprocess.run([
            sys.executable, '-m', 'zeusdl',
            '--cookies', COOKIES_PATH,
            '-f', f'bestvideo[height<={QUALITY}]+bestaudio/best[height<={QUALITY}]/best',
            '--merge-output-format', 'mp4',
            '-o', OUTPUT_DIR + '/%(title)s.%(ext)s',
            '--concurrent-fragments', str(WORKERS),
            args
        ])
        status = 'done' if result.returncode == 0 else 'error'
        print(f'[Agent] Download {status}')
        report_status(status)

    elif cmd_name == 'push' and 'telegram' in args:
        send_to_telegram(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
        report_status('done')

    else:
        print(f'[Agent] Commande inconnue: {cmd_name}')
        report_status('ready')

# --- Demarrer ---
if register():
    print(f'Agent {AGENT_ID} connecte au Mastermind')
    print(f'Depuis @TGBOX_HQbot envoie :')
    print(f'  /agents')
    print(f'  /send {AGENT_ID} download https://site-ma.bangbros.com/scenes?addon=5971')
    print(f'  /send {AGENT_ID} push telegram')
    print('En attente d\'ordres...')
    while True:
        order = poll_order()
        if order:
            threading.Thread(target=execute, args=(order,), daemon=True).start()
        time.sleep(3)
else:
    print('Mastermind non joignable. Lance les cellules 4 et 5 manuellement.')
```
