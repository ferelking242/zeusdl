```python
# CELLULE 5 - Envoyer vers Telegram
import os, requests

def send_to_telegram(folder, bot_token, channel):
    if not channel:
        print('CHANNEL vide - mets ton channel ID dans la cellule 2')
        return
    exts = ('.mp4', '.mkv', '.avi', '.mov', '.webm')
    files_list = [f for f in os.listdir(folder) if f.lower().endswith(exts)]
    if not files_list:
        print('Aucune video dans', folder)
        return
    print(f'{len(files_list)} video(s) a envoyer vers {channel}')
    for fname in files_list:
        path = os.path.join(folder, fname)
        size_mb = os.path.getsize(path) / 1024 / 1024
        if size_mb > 2000:
            print(f'Ignore (trop gros {size_mb:.0f}MB): {fname}')
            continue
        print(f'Envoi {fname} ({size_mb:.1f}MB)...')
        with open(path, 'rb') as f:
            r = requests.post(
                f'https://api.telegram.org/bot{bot_token}/sendVideo',
                data={'chat_id': channel, 'caption': fname[:1000]},
                files={'video': (fname, f, 'video/mp4')},
                timeout=300
            )
        print('  OK' if r.status_code == 200 else f'  Erreur: {r.text[:100]}')

send_to_telegram(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
```

```python
# CELLULE 6 - Agent Hermes (optionnel)
# Lance cette cellule pour etre controle depuis @TGBOX_HQbot
import time, threading, requests as req

class Agent:
    def __init__(self):
        self.url = MASTER_URL.rstrip('/')
        self.running = True

    def register(self):
        try:
            r = req.post(f'{self.url}/register', json={'agent_id': AGENT_ID, 'status': 'ready'}, timeout=5)
            return r.status_code == 200
        except Exception as e:
            print('Mastermind inaccessible:', e)
            return False

    def poll(self):
        try:
            r = req.get(f'{self.url}/orders/{AGENT_ID}', timeout=5)
            if r.status_code == 200:
                return r.json().get('orders', [])
        except Exception:
            pass
        return []

    def ack(self, oid):
        try:
            req.post(f'{self.url}/ack', json={'agent_id': AGENT_ID, 'order_id': oid}, timeout=5)
        except Exception:
            pass

    def execute(self, order):
        import subprocess
        parts = order.strip().split(None, 1)
        cmd_name = parts[0].lower()
        args = parts[1] if len(parts) > 1 else ''
        print(f'[Agent] {order}')
        if cmd_name == 'download' and args:
            subprocess.run(['yt-dlp', '--cookies', COOKIES_PATH,
                '-f', f'bestvideo[height<={QUALITY}]+bestaudio/best[height<={QUALITY}]',
                '--merge-output-format', 'mp4',
                '-o', OUTPUT_DIR + '/%(title)s.%(ext)s',
                '--concurrent-fragments', str(WORKERS), args])
        elif cmd_name == 'push' and 'telegram' in args:
            send_to_telegram(OUTPUT_DIR, BOT_TOKEN, CHANNEL)
        else:
            print('[Agent] Commande inconnue:', cmd_name)

    def start(self):
        if not self.register():
            print('Lance les cellules 4 et 5 directement.')
            return
        print(f'Agent {AGENT_ID} connecte')
        print(f'Depuis @TGBOX_HQbot :')
        print(f'  /agents')
        print(f'  /send {AGENT_ID} download https://site-ma.bangbros.com/scenes?addon=5971')
        print(f'  /send {AGENT_ID} push telegram')
        while self.running:
            try:
                for o in self.poll():
                    oid = o.get('id')
                    text = o.get('command') or o.get('order') or ''
                    if text:
                        threading.Thread(target=self.execute, args=(text,), daemon=True).start()
                    if oid:
                        self.ack(oid)
            except Exception as e:
                print('[Agent] Erreur:', e)
            time.sleep(3)

Agent().start()
```
